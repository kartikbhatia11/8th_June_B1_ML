```python
import pandas as pd
import numpy as np
import random
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
import tkinter as tk
from tkinter import ttk
```

```python
user_data = pd.read_csv('ml-100k/u.data', sep = '\t', names = ["user id","movie id","rating","timestamp"])

cols = ("movie id | movie title | release date | video release date | IMDb URL | unknown | Action | Adventure | Animation | Children's | Comedy | Crime | Documentary | Drama | Fantasy | Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi | Thriller | War | Western").split(" | ")
movies_data = pd.read_csv("ml-100k/u.item", sep = '|', names = cols)

col = ['movie id', 'movie title',  'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir','Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War','Western']
Data = pd.merge(user_data[["user id","movie id","rating"]], movies_data[col], on = 'movie id')

avg_rating = Data.groupby('movie title')['rating'].mean()

rating_count = Data.groupby('movie title')['rating'].count()

Data.drop('movie id', axis=1, inplace=True)



    
# GUI Code

root = tk.Tk()
root.title('Movie Recommendation System')
root.geometry('700x510')
root.config(bg = 'sky blue')

list_b = tk.Listbox(root, width = 40, height = 30, background = 'light slate grey',font = ['Arial',10,'bold'])
list_b.place(x=10,y=10, bordermode = 'inside')

for i,t in enumerate(movies_data['movie title'].unique()) :
    list_b.insert(i+1,t)


def recommend():
    user = list_b.get(list_b.curselection())
    
    Data_pivot = Data[Data['movie title'] == user]

    genres = ['unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir','Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War','Western']
    m_genre = []
    for i in genres:
        if Data_pivot[i].head(1).to_list()[0] == 1 :
            m_genre.append(i)
    m_genre

    d = Data[Data[random.choice(m_genre)] == 1]

    Rating_pivot = d.pivot_table(index = 'user id', columns = 'movie title', values = 'rating')
    movie_corr = Rating_pivot.corrwith(Rating_pivot[user])
    movie_corr = pd.DataFrame(movie_corr, columns = ['Correlation'])
    movie_corr['Count'] = rating_count
    movie_corr['Avg Rating'] = avg_rating

    rec = movie_corr[movie_corr['Count'] > 100].sort_values(by='Correlation', ascending = False)
    recommend = rec.index[1:6]
    
    res1.set(recommend[0])
    res2.set(recommend[1])
    res3.set(recommend[2])
    res4.set(recommend[3])
    res5.set(recommend[4])

    
colors = ['papaya whip', 'blanched almond', 'bisque', 'peach puff',
    'navajo white', 'lemon chiffon','dark khaki', 'khaki', 'pale goldenrod', 'light goldenrod yellow',
    'light yellow','LemonChiffon2', 'LemonChiffon3', 'LemonChiffon4', 'cornsilk2', 'cornsilk3',
    'cornsilk4','SeaGreen1', 'SeaGreen2', 'SeaGreen3', 'PaleGreen1', 'PaleGreen2',
    'PaleGreen3', 'PaleGreen4', 'SpringGreen2', 'SpringGreen3', 'SpringGreen4','khaki1', 'khaki2', 
    'khaki3', 'khaki4','LightGoldenrod1', 'LightGoldenrod2', 'LightGoldenrod3', 'LightGoldenrod4',
    'sienna1', 'sienna2', 'sienna3', 'sienna4', 'burlywood1','burlywood2', 'burlywood3', 'burlywood4', 
    'wheat1', 'wheat2', 'wheat3', 'wheat4', 'tan1','tan2', 'tan4',]
    
tk.Button(root, text = "Recommend", bg = 'orange2', borderwidth = 2, padx = 10, pady=20, relief = 'raised', command = recommend, width = 15, font = ['Arial',15,'bold']).place(x = 400, y = 20)

res1 = tk.Variable(root)
res1.set('')
tk.Label(root, textvariable = res1, bg = random.choice(colors), font = ['Arial',10,'bold','italic']).place(x = 330, y = 150)

res2 = tk.Variable(root)
res2.set('')
tk.Label(root, textvariable = res2, bg = random.choice(colors), font = ['Arial',10,'bold','italic']).place(x = 330, y = 200)

res3 = tk.Variable(root)
res3.set('')
tk.Label(root, textvariable = res3, bg = random.choice(colors), font = ['Arial',10,'bold','italic']).place(x = 330, y = 250)

res4 = tk.Variable(root)
res4.set('')
tk.Label(root, textvariable = res4, bg = random.choice(colors), font = ['Arial',10,'bold','italic']).place(x = 330, y = 300)

res5 = tk.Variable(root)
res5.set('')
tk.Label(root, textvariable = res5, bg = random.choice(colors), font = ['Arial',10,'bold','italic']).place(x = 330, y = 350)

root.mainloop()
```
